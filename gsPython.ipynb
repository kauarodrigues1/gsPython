```python
solar_consumo = [150, 160, 155, 158, 145, 152, 147, 155, 160, 150] 
solar_custo = [80, 82, 79, 81, 80, 83, 82, 81, 84, 80]

eolica_consumo = [200, 210, 220, 215, 210, 205, 218, 212, 220, 205]
eolica_custo = [60, 61, 61, 63, 60, 59, 62, 63, 61, 60]

hid_consumo = [500, 510, 505, 495, 480, 490, 500, 505, 510, 495]
hid_custo = [50, 51, 50, 49, 48, 47, 50, 51, 50, 49]


def soma_consumo():

    solar_consumo_total = sum(solar_consumo)
    eolica_consumo_total = sum(eolica_consumo)
    hid_consumo_total = sum(hid_consumo)

    return solar_consumo_total, eolica_consumo_total, hid_consumo_total

solar_consumo_total, eolica_consumo_total, hid_consumo_total = soma_consumo()

print("-"*40)
print('ENERGIA SOLAR')
print(f"Soma do consumo total: {solar_consumo_total} MW ")
print("-"*40)

print('ENERGIA EOLICA')
print(f"Soma do consumo total: {eolica_consumo_total} MW ")
print("-"*40)

print('ENERGIA HIDRELETRICA')
print(f"Soma do consumo total: {hid_consumo_total} MW ")
print("-"*40)
```

```python
def soma_custo():
    mult_solar = 0
    mult_eolica = 0
    mult_hid = 0

    for num1, num2 in zip(solar_custo, solar_consumo):
        mult_solar += num1* num2 
        

    for num1, num2 in zip(eolica_custo, eolica_consumo):
        mult_eolica += num1*num2

    for num1, num2 in zip(hid_custo, hid_consumo):
        mult_hid += num1*num2

    return mult_solar, mult_eolica, mult_hid
          

mult_solar, mult_eolica, mult_hid = soma_custo()

print("-"*40)
print('ENERGIA SOLAR')
print(f"Soma do custo total: ${mult_solar} ")
print("-"*40)

print('ENERGIA EOLICA')
print(f"Soma do custo total: ${mult_eolica} ")
print("-"*40)

print('ENERGIA HIDRELETRICA')
print(f"Soma do custo total: ${mult_hid} ")
print("-"*40)
```

```python
solar_custo_total_for = 0
solar_consumo_total = sum(solar_consumo)

eolica_custo_total_for = 0
eolica_consumo_total = sum(eolica_consumo)

hid_custo_total_for = 0
hid_consumo_total = sum(hid_consumo)

# media_custo_solar ////////////////////////

for num1, num2 in zip(solar_custo, solar_consumo):
    solar_custo_total_for += num1* num2 
media_custo_solar = solar_custo_total_for / solar_consumo_total

for num1, num2 in zip(eolica_custo, eolica_consumo):
    eolica_custo_total_for += num1* num2 
media_custo_eolica = eolica_custo_total_for / eolica_consumo_total

for num1, num2 in zip(hid_custo, hid_consumo):
    hid_custo_total_for += num1* num2 
media_custo_hid = hid_custo_total_for / hid_consumo_total


menor_custo = min(media_custo_solar, media_custo_eolica, media_custo_hid)

if menor_custo == media_custo_solar:
    fonte_menor_custo = "Solar"
elif menor_custo == media_custo_eolica:
    fonte_menor_custo = "Eólica"
else:
    fonte_menor_custo = "Hidrelétrica"

maior_custo = max(media_custo_solar, media_custo_eolica, media_custo_hid)

if maior_custo == media_custo_solar:
    fonte_maior_custo = "Solar"
elif maior_custo == media_custo_eolica:
    fonte_maior_custo = "Eólica"
else:
    fonte_maior_custo = "Hidrelétrica"


print('-'*40)
print('ENERGIA SOLAR')
print(f'Custo medio é: ${media_custo_solar:.2f}')
print('-'*40)
print('ENERGIA EOLICA')
print(f'Custo medio é: ${media_custo_eolica:.2f}')
print('-'*40)
print('ENERGIA HIDRELETRICA')
print(f'Custo media é: ${media_custo_hid:.2f}')
print('-'*40)
print('COMARAÇÃO')
print(f'Menor média de custo: {fonte_menor_custo}')
print(f'Maior média de custo: {fonte_maior_custo}')
print('-'*40)
```
